```python
with open('data/EBM_yan_section_filtered/output/fold1/output_update_256/microsoft-BiomedNLP-PubMedBERT-base-uncased-abstract/predict_test_label.txt','r') as f:
    lines = f.read().splitlines()
    print (lines)
```

```python
y_true = [line.split('\t')[1] for line in lines if len(line.split('\t'))==3]
y_pred = [line.split('\t')[2] for line in lines if len(line.split('\t'))==3]
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt 
import numpy as np

labels = list(set(y_true+y_pred))
cm = confusion_matrix(y_true, y_pred, labels,normalize= None)
cmn = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
cmn = np.round(cmn, 2)

cmd = ConfusionMatrixDisplay(cmn, display_labels=labels)

#ax.set_xticklabels(labels, rotation=45, ha='right', rotation_mode='anchor')
fig, ax = plt.subplots(figsize=(9,9),dpi=600)

cmd.plot(ax=ax)
plt.setp(ax.get_xticklabels(),ha="right", rotation_mode="anchor")

plt.show()
```
